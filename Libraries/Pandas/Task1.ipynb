```python
#Reshape Data: Concatenate
import pandas as pd
df1 = pd.DataFrame({
    "student_id": [1, 2, 3, 4],
    "name": ["Mason", "Ava", "Taylor", "Georgia"],
    "age": [8, 6, 15, 17]
})
df2 = pd.DataFrame({
    "student_id": [5, 6],
    "name": ["Leo", "Alex"],
    "age": [7, 7]
})
result = pd.concat([df1, df2], ignore_index=True)
print(result)
```

```python
#Reshape Data: Pivot
import pandas as pd
weather = pd.DataFrame({
    "city": [
        "Jacksonville", "Jacksonville", "Jacksonville", "Jacksonville", "Jacksonville",
        "ElPaso", "ElPaso", "ElPaso", "ElPaso", "ElPaso"],
    "month": [
        "January", "February", "March", "April", "May",
        "January", "February", "March", "April", "May"],
    "temperature": [13, 23, 38, 5, 34, 20, 6, 26, 2, 43]
})
result = weather.pivot(index='month', columns='city', values='temperature').reset_index()
print(result)
```

```python
#Reshape Data: Melt
import pandas as pd
report = pd.DataFrame({
    "product": ["Umbrella", "SleepingBag"],
    "quarter_1": [417, 800],
    "quarter_2": [224, 936],
    "quarter_3": [379, 93],
    "quarter_4": [611, 875]
})
result = report.melt(
    id_vars="product",
    var_name="quarter",
    value_name="sales"
)
print(result)
```

```python
#Method chaining
import pandas as pd
animals = pd.DataFrame({
    "name": ["Tatiana", "Khaled", "Alex", "Jonathan", "Stefan", "Tommy"],
    "species": ["Snake", "Giraffe", "Leopard", "Monkey", "Bear", "Panda"],
    "age": [98, 50, 6, 45, 100, 26],
    "weight": [464, 41, 328, 463, 50, 349]
})
result = (animals[animals['weight'] > 100].sort_values('weight', ascending=False)[['name']])
print(result.to_string(index=False))
```

```python
                                      # frist problem
# Costomer who Never Order
import pandas as pd
customers = pd.DataFrame({
    "id": [1, 2, 3, 4],
    "name": ["Joe", "Henry", "Sam", "Max"]
})
orders = pd.DataFrame({
    "id": [1, 2],
    "customerId": [3, 1]
})
merged = customers.merge(orders, left_on="id", right_on="customerId", how="left")
no_orders = merged[merged["id_y"].isna()]["name"]
result = pd.DataFrame({"Customers": no_orders})
print(result)
```

```python
                              
                             #SECOND PROBLEM
import pandas as pd
data = {
    "employee_id": [2, 3, 7, 8, 9],
    "name": ["Meir", "Michael", "Addilyn", "Juan", "Kannon"],
    "salary": [3000, 3800, 7400, 6100, 7700]
}
df = pd.DataFrame(data)
df["bonus"] = 0 
df["salary"]=df.loc[ (df["employee_id"] % 2 != 0) & (df["name"].str.startswith("M")),"bonus"] 
result = df[["employee_id", "bonus"]].sort_values(by="employee_id")
print(result)
```

```python
                                             #THIRD PROBLEM
import pandas as pd
data = {
    "id": [1],
    "salary": [100]
}
df = pd.DataFrame(data)
salaries = df["salary"].drop_duplicates().sort_values(ascending=False)
second_highest = salaries.iloc[1] if len(salaries) > 1 else None
result = pd.DataFrame({"SecondHighestSalary": [second_highest]})
print(result)
```

```python
import pandas as pd
import numpy as np

data = {
    "account_id": [3, 2, 8, 6],
    "income": [108939, 12747, 87709, 91796]
}
df = pd.DataFrame(data)
conditions = [df["income"] < 20000,(df["income"] >= 20000) & (df["income"] <= 50000),df["income"] > 50000]
choices = ["Low Salary", "Average Salary", "High Salary"]
df["category"] = np.select(conditions, choices)
```

```python
                                                    #PROBLEM five 
import pandas as pd
data = {
    "player_id": [1, 1, 2, 3, 3],
    "device_id": [2, 2, 3, 1, 4],
    "event_date": ["2016-03-01", "2016-05-02", "2017-06-25", "2016-03-02", "2018-07-03"],
    "games_played": [5, 6, 1, 0, 5]
}

df = pd.DataFrame(data)
df["event_date"] = pd.to_datetime(df["event_date"])
result = df.groupby("player_id").agg(first_login=("event_date", "min")).reset_index()
result.columns = ["player_id", "first_login"]
print(result)
```
